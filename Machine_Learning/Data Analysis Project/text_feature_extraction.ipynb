```python
import numpy as np
import pandas as pd

def your_code_here():
    return 0

df_train = pd.read_csv("train.csv")
df_test  = pd.read_csv("test.csv")
df_dev   = pd.read_csv("dev.csv")

df_train
```

```python
from sklearn.feature_extraction.text import CountVectorizer
counter = CountVectorizer()
counter
```

```python
df_train_text = df_train['text']

X_train = counter.fit_transform(df_train_text)
analyze = counter.build_analyzer()
cols = counter.get_feature_names().copy()
cols
```

```python
X_test = counter.transform(df_test['text'])
(X_train,X_test)
```

## Transform Vectorized Text Dataframe into tfidf matrix

```python
from sklearn.feature_extraction.text import TfidfTransformer as tf
transformer = tf(smooth_idf=False)
transformer
```

```python
tfidf = transformer.fit_transform(X_train.toarray())
tfidf.toarray()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer as tfv
vectorizer = tfv()
X_train_pp = vectorizer.fit_transform(df_train_text)
X_test_pp  = vectorizer.transform(df_test['text'])
df_train_pp = pd.DataFrame(data=X_train_pp.toarray(),columns = cols)
df_test_pp = pd.DataFrame(data=X_test_pp.toarray(), columns = cols)
```

```python
train = pd.concat([df_train_pp,df_train[['label','category']]],1)
test  = pd.concat([df_test_pp, df_dev[['label','category']]])
```

```python
train.to_csv("train_pp.csv")
test.to_csv("test_pp.csv")
```

```python
train.replace(['NOT', 'OFF', 'NaN', 'UNT', 'TIN'], [0, 1, 0, 1, 2])
test.replace(['NOT', 'OFF', 'NaN', 'UNT', 'TIN'], [0, 1, 0, 1, 2])
```
